```python
import scipy
import librosa
import os
import csv
import csv
import numpy as np
import pandas as pd
from scipy.stats import skew
from scipy.stats import kurtosis
```

Creez un dataframe gol cu numele coloanelor.

```python
df = pd.DataFrame(columns=['file_name', 'duration', 'mean','variance','skewness','kurtosis','label'])
```

```python
controls = "Control Patologicas".split()
vowels = "A E I O U".split()
```

Extrag mean, variance, skewness and kurtosis pentru fiecare audio.

```python
for c in controls:
  for v in vowels:
    for filename in os.listdir(f'/content/drive/MyDrive/Audio Data/Vowels/{c}/{v}'):      #change here directory location depending on where you run !!!
      patient_code = filename[:-7]

      recordingname = f'/content/drive/MyDrive/Audio Data/Vowels/{c}/{v}/{filename}'
      y, sr = librosa.load(recordingname, mono=True)
      duration = librosa.get_duration(y=y, sr=sr)
      mean = np.mean(y)
      variance = np.var(y)
      # Calculate the skewness
      skeweness = skew(y, axis=0, bias=True)
      # Calculate the kurtosis
      k = kurtosis(y, axis=0, bias=True)
      df = df.append({'file_name': filename, 'duration': duration, 'mean' : mean,'variance':variance,'skewness':skeweness,'kurtosis':k,'label':c}, ignore_index=True)
```

  Salvez dataframe într-un csv.

```python
with open('/content/drive/MyDrive/Parkinson CSV files/allAudios_MeanVarianceSkewnessKurtosis.csv', 'w') as csv_file:
    df.to_csv(path_or_buf=csv_file)
```

```python

import plotly.graph_objects as go

import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Parkinson CSV files/allAudios_MeanVarianceSkewnessKurtosis.csv")
```

```python
fig = go.Figure()


fig.add_trace(go.Violin(
                        y=df['mean'][ df['label'] == 'Control' ],
                        legendgroup='Control', scalegroup='Control', name='Control',
                        line_color='blue')
             )
fig.add_trace(go.Violin(
                        y=df['mean'][ df['label'] == 'Patologicas' ],
                        legendgroup='Patologicas', scalegroup='Patologicas', name='Patologicas',
                        line_color='orange')
             )
fig.update_traces(meanline_visible=True)
fig.update_layout(violingap=0, violinmode='overlay')
fig.update_layout(title_text='Mean')
fig.show()
```
